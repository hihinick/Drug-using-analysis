# 資料讀進來後，先對 '藥品ATC code' 進行編碼，編碼順序為 ATC_name_seq (list型態)

```python
labelencoder = LabelEncoder()
data['ATC encode'] = labelencoder.fit_transform(data['藥品ATC code'])
data.head(20)
```

```python
#取得 matrix的columns的順序(label encoding的順序) --> ATC_name_seq
ATC_max = data['ATC encode'].max()
ATC_name_seq = []
for i in range(ATC_max+1):
    filter1 = data['ATC encode'] == i
    aaa = data[filter1] 
    # print(i,aaa)
    aaa.reset_index(drop=True, inplace=True)
    ATC_name_seq.append(aaa['藥品ATC code'][0])
len(ATC_name_seq)
```

# 同個病例號做group by

```python
def patient_groupby_v1(data):
    one_ = data.groupby(by='病例號',axis=0)
    one_ = list(one_)
    patient_id = []
    disease = []
    ATC_code = []
    for i in range(len(one_)):
        one = one_[i][1]
        # id
        patient_id.append(one_[i][0])
        # disease
        save = list(one.values[0][1:8])
        save = [n for n in save if n != 'na' ]
        disease.append(save)
        # ATC
        ATC_code.append(list(one['ATC encode'].unique()))
    form = pd.DataFrame()
    form['patient_id'] = patient_id
    form['disease'] = disease
    form['ATC_code'] = ATC_code
    return form, patient_id, disease, ATC_code 
```

```python
data
```

```python
groupby = patient_groupby_v1(data)
Patient_groupby = groupby[0]
patient_id = groupby[1]
disease = groupby[2]
ATC_code = groupby[3]
```

```python
# input
Patient_groupby
```

```python
# 編碼
def to_matrix(data, col):
    for i in range(len(data)):
        data[i] = keras.utils.np_utils.to_categorical(np.array(data[i]), num_classes=col)
        data[i] = np.sum(data[i], axis=0)
        # data[i] = data[i].astype(bool)
    return data
```

```python
vector = to_matrix(ATC_code, len(ATC_name_seq))
Patient_groupby['to_matrix'] = vector
# output
Patient_groupby
```

```python
# 計算每個patient_id被診斷出的疾病數量跟用藥品項數量
print(Patient_groupby['ATC_code'][0],len(Patient_groupby['ATC_code'][0]))
disease_counts = []
drug_counts = []
for i in range(len(Patient_groupby)):
    c = len(Patient_groupby['ATC_code'][i])
    cc = len(Patient_groupby['disease'][i][1:])
    drug_counts.append(c)
    disease_counts.append(cc)
Patient_groupby['disease_counts'] = disease_counts
Patient_groupby['drug_counts'] = drug_counts
Patient_groupby = Patient_groupby.sort_values(by='drug_counts', ascending=False)
Patient_groupby.reset_index(drop=True)
```

```python
# 將疾病從	patient_id	分出來
disease_bom = []
to_matrix_bom = []
for i in range(len(Patient_groupby)):
    target = Patient_groupby['disease'][i]
    for a in range(1,len(target)):
        disease_bom.append(target[a])
        to_matrix_bom.append(Patient_groupby['to_matrix'][i])

form1 = pd.DataFrame()
form1['disease_bom'] = disease_bom
form1['ATC_code_bom'] = to_matrix_bom
form1
```

```python
input_apriori = list(form1.groupby(by='disease_bom',axis=0))
len(input_apriori)
```
